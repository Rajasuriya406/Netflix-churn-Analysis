```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('F:/netflix-churn/netflix_customer_churn.csv')
```

```python
df.head()
df.info()
df.duplicated().sum()
```

```python
numeric_cols = ["age", "watch_hours", "last_login_days", "monthly_fee", "avg_watch_time_per_day"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
categorical_cols = ["gender", "subscription_type", "region", "device", "payment_method", "favorite_genre"]
df[categorical_cols] = df[categorical_cols].astype("category")
```

```python
# handling invalid values for age 
df = df[(df["age"] >= 18) & (df["age"] <= 90)]

#watch hours

df=df[df["watch_hours"]>=0]

```

```python
#Encoding churn
df["churned"]=df["churned"].astype(int)
```

```python
#standardizing text
df["region"]=df["region"].str.strip().str.title()
df["device"]=df["device"].str.strip().str.title()
df["payment_method"]=df["payment_method"].str.strip().str.title()
```

```python
#outier treatment
q1=df["watch_hours"].quantile(0.25)
q3=df["watch_hours"].quantile(0.75)
IQR=q3-q1

df=df[(df["watch_hours"]>=q1-1.5*IQR)&(df["watch_hours"]<=q3+1.5*IQR)]
```

```python
df.describe()
df.isnull().sum()
df.info()
```

```python
df.to_csv("netflix_customer_churn.csv", index=False)
```
